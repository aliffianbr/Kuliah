```python
# Mendefinisikan fungsi bubble_sort untuk mengurutkan array menggunakan Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Mendefinisikan fungsi selection_sort untuk mengurutkan array menggunakan Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

# Mendefinisikan fungsi insertion_sort untuk mengurutkan array menggunakan Insertion Sort
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Mengambil input dari pengguna dan mengonversi ke dalam list angka
numbers = list(map(int, input("Masukkan angka yang akan diurutkan, pisahkan dengan spasi: ").split()))

# Menggunakan Bubble Sort
bubble_sorted = numbers.copy()
bubble_sort(bubble_sorted)
print(f'Bubble Sort: {bubble_sorted}')

# Menggunakan Selection Sort
selection_sorted = numbers.copy()
selection_sort(selection_sorted)
print(f'Selection Sort: {selection_sorted}')

# Menggunakan Insertion Sort
insertion_sorted = numbers.copy()
insertion_sort(insertion_sorted)
print(f'Insertion Sort: {insertion_sorted}')
```
