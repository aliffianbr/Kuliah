Program mahasiswa menggunakan Dictionary



```python
# Import library python
import datetime  			# Mengimpor modul datetime untuk bekerja dengan tanggal dan waktu
import os  						# Mengimpor modul os untuk berinteraksi dengan sistem operasi
import string  				# Mengimpor modul string untuk operasi pada string
import random  				# Mengimpor modul random untuk menghasilkan bilangan acak

# template dictionary mahasiswa
mahasiswa_template = {  	# Membuat template dictionary untuk data mahasiswa
    'nama': 'nama',  			# Kunci 'nama' dengan nilai awal 'nama'
    'nim': '00000000',  	# Kunci 'nim' dengan nilai awal '00000000'
    'sks_lulus': 0,  			# Kunci 'sks_lulus' dengan nilai awal 0
    'lahir': datetime.datetime(1111, 1, 11)  			# Kunci 'lahir' dengan nilai datetime 1111-01-11
}

data_mahasiswa = {}  # Membuat kamus kosong untuk menyimpan data mahasiswa

while True: 	 # Loop akan berjalan selama kondisinya benar (selamanya)
    # os.system("cls")		 	# untuk windows
    os.system("clear")  		# Membersihkan layar (dapat berbeda tergantung pada sistem operasi)
    print(f"{'SELAMAT DATANG':^20}") 	 # Mencetak teks SELAMAT DATANG ditengah dengan lebar 20 karakter
    print(f"{'DATA MAHASISWA USM':^20}") 	 # Mencetak teks DATA MAHASISWA ditengah dengan lebar 20 karakter
    print("-" * 20)  	# Mencetak garis pemisah sebanyak 20 karakter

    # Membuat dictionary baru untuk data mahasiswa sesuai template
    mahasiswa = dict.fromkeys(mahasiswa_template.keys())

    # Meminta pengguna untuk memasukkan data mahasiswa
    mahasiswa['nama'] = input("Nama Mahasiswa: ")
    mahasiswa['nim'] = input("NIM Mahasiswa: ")
    mahasiswa['sks_lulus'] = int(input("SKS Lulus: "))
    TAHUN_LAHIR = int(input("Tahun lahir (YYYY): "))
    BULAN_LAHIR = int(input("Bulan lahir (1-12): "))
    TANGGAL_LAHIR = int(input("Tanggal lahir (1-31): "))
    mahasiswa['lahir'] = datetime.datetime(TAHUN_LAHIR, BULAN_LAHIR, TANGGAL_LAHIR)

    KEY = ''.join((random.choice(string.ascii_uppercase) for i in range(6)))  # Membuat kunci acak sepanjang 6 karakter
    data_mahasiswa.update({KEY: mahasiswa})  # Menambahkan data mahasiswa ke dalam kamus menggunakan kunci acak

    # Menghilangkan beasiswa
    print(f"\n{'KEY':<6} {'Nama':<17} {'NIM':<8} {'SKS Lulus':<10} {'Tanggal Lahir':<10}")  # Mencetak header tabel

    for mahasiswa in data_mahasiswa:  # Loop untuk setiap data mahasiswa dalam kamus
        KEY = mahasiswa 	 # Menetapkan kunci mahasiswa saat ini ke variabel KEY

        # Mendapatkan nilai untuk setiap atribut mahasiswa
        NAMA = data_mahasiswa[KEY]['nama']
        NIM = data_mahasiswa[KEY]['nim']
        SKS = data_mahasiswa[KEY]['sks_lulus']
        LAHIR = data_mahasiswa[KEY]['lahir'].strftime("%x")  # Format tanggal lahir

        # Mencetak data mahasiswa dalam format yang telah ditentukan
        print(f"{KEY:<6} {NAMA:<17} {NIM:<8} {SKS:^10} {LAHIR:^10}")

    print("\n")  # Memberikan baris kosong

    is_done = input("Ingin input mahasiswa lagi (y/n)? ")  # Meminta masukan untuk melanjutkan atau tidak
    if is_done == "n":  # Jika masukan adalah 'n'
        break  # Keluar dari loop
```
