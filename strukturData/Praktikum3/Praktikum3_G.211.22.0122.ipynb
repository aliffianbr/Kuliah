```python
# Membuat data list
data_list = ['p', 'r', 'o', 'g', 'r', 'a', 'm', 'm', 'i', 'n', 'g']

# Variabel untuk menyimpan hasil
output_list = ""

# Mengonstruksi kata "programming" dengan perulangan
for karakter in data_list:
    output_list += karakter

print(output_list)  # Hasil: "programming"
```
