```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataset = pd.read_csv("DataStock.csv")
```

```python
dataset.info()
```

```python
kolom = ['Purchase Price', 'Current Price', 'Gain/Loss']
for i in kolom:
    data = dataset[i].str.replace(',', '').str.replace('$', '').astype(float)
    
    data.plot.hist(alpha=1, color='blue', bins=100)
    plt.ylim(0,8)
    plt.xlim(min(data), max(data))
    plt.xlabel(i)
    plt.ylabel('Frekuensi')
    plt.title(f'Distribusi {i}')
    plt.rcParams['figure.figsize'] = [6, 4]
    plt.rcParams['figure.dpi'] = 150
    plt.show()
```

```python
dataset = pd.read_csv("DataStock.csv", index_col=2)
dataset['Dividend/Share'] = dataset['Dividend/Share'].str.replace('$', '').astype(float)
dividendTertinggiStockName = dataset.loc["AAPL":"TWTR", ("Stock Name", "Dividend/Share")]
dividendTertinggiStockName = dividendTertinggiStockName.nlargest(5, 'Dividend/Share')
stockName = dividendTertinggiStockName['Stock Name']
dividend = dividendTertinggiStockName['Dividend/Share']
plt.bar(stockName[:], dividend[:])
plt.title("Lima Dividend Tertinggi")
plt.xlabel("Stock Name")
plt.ylabel("Dividend/Shares")
plt.show()
```
